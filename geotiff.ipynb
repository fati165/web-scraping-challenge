```python
import requests
from bs4 import BeautifulSoup
import csv
import urllib
```

```python
URL_part1 = "https://data.humdata.org/organization/facebook?res_format=GeoTIFF&page="
```

```python
URL_part3 = "&q=&ext_page_size=25#datasets-section"
```

```python
URL_part2 = 1
while True:
    URL = URL_part1+str(URL_part2)+URL_part3
    print ('URL: '+URL)
    r = requests.get(URL)
    soup = BeautifulSoup(r.content, 'html5lib') 
    #print(soup.prettify())
    table = soup.find('div', attrs = {'id':'search-page-results'})
    #print(table.prettify())
    noItemTable = soup.find('h2')
    #print(noItemTable is not None)
    if(noItemTable is not None):
        noItemTableText = noItemTable.text
        #print('"'+noItemTableText+'"')
    if(noItemTable is not None and noItemTableText=='Sorry no datasets found for current filter selection '):
        break;
    URL_part2= URL_part2+1

    for row in table.findAll('a', attrs = {'class':'label label-small', 'data-format':'geotiff'}):
          #print(row)
        country_est = row['href']
          #print(country_est)
        secondSlash = country_est.find('/', 1)
          firstHiphen = country_est.find('-high-resolution-population', 9)
          #print(firstHiphen)
          if(firstHiphen>-1):
            dirName = country_est[secondSlash+1:firstHiphen]
        else:
            dirName = country_est[secondSlash+1:]
          #print(dirName)
          country_href = 'https://data.humdata.org/'+row['href']
          #print(country_href)
          #print()
        
          rTiff = requests.get(country_href)
        tableTiff = BeautifulSoup(rTiff.content, 'html5lib') 
          for rowTiff in tableTiff.findAll('a', attrs = {'class':'btn btn-empty btn-empty-blue hdx-btn resource-url-analytics ga-download', 'title':'Download'}):
            #print(rowTiff)
            downloadLink = rowTiff['href']
            #print(downloadLink)
            if(downloadLink.find('geotiff')==-1 or downloadLink.find('general')==-1):
            continue
            fileName = downloadLink[downloadLink.rfind('/')+1:]
            downloadLinkTotal = 'https://data.humdata.org'+downloadLink
            print(fileName+' '+downloadLinkTotal)
            print()
            urllib.request.urlretrieve(downloadLinkTotal, fileName) 
          #break
```
